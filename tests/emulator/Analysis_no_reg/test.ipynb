```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(2, 8):
    p = np.load(str(i)+"_order_0.npy")
    print(i , len(p))
    plt.plot(np.arange(len(p)), p , label = str(i)+" order")
plt.yscale("log")
plt.title("Anderson Acc. for different orders with regression term $\lambda = 0$")
plt.legend()
plt.ylabel("max_diff")
plt.xlabel("iteration_step")
plt.show()    
```
