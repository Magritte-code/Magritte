```python
#Esc->R->Y

import numpy as np
import matplotlib.pyplot as plt
#numbers = ['2','1','0.1','0.01','0.001', '0.0001' 
#           ,'1e-07','1e-08','1e-09','1e-10' ,
#           '1e-11','1e-12','1e-17','1e-20','0']
numbers = ['2','1', '0.0001' 
            ,
           '1e-12','1e-20','0']
m=2
for i in numbers:
    print(i)
    arr = np.load(str(m)+'_order_'+str(i)+".npy")
    plt.plot(np.arange(len(arr)),arr,'-' , label = 'l='+i)
    plt.yscale('log')
plt.legend()
plt.show()

```

```python
for p in numbers:
    arr = np.load("C_"+str(m)+'order_'+(p)+".npy")
    c = []
    for i in range(0, len(arr)-1):
        t = arr[i+1] - arr[i]
        t = np.sum(t)
        t = t**2
        c.append(t)
    #print(np.arange(len(c)))
    #plt.plot(np.arange(len(c)), c , 'ro')
    plt.stem(np.arange(len(c)),c , "ro" , label = "l = "+p )
    plt.yscale("log")
    plt.legend()
    plt.show()
```

```python
for p in numbers:
    l_norm = []
    arr = np.load("C_"+str(m)+'order_'+(p)+".npy" )
    for i in arr:
        l_norm.append((i[0] - i[1] - i[2])**2)
    plt.stem(np.arange(len(l_norm)) , l_norm ,  label = 'l = '+p)
    plt.yscale('log')
    plt.legend()
    plt.show()
```
