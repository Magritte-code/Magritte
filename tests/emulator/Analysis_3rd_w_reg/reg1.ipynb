```python
import numpy as np
import matplotlib.pyplot as plt
numbers = [ '0.0001' ,'1e-06','1e-08','1e-10' , '1e-12','1e-14' , '1e-16' , '1e-18','0']
m=3
for i in numbers:
    print(i)
    arr = np.load(str(m)+'_order_'+str(i)+".npy")
    plt.plot(np.arange(len(arr)),arr,label = 'l = '+i +" and iteration = " + str(len(arr)))
    plt.yscale('log')
    plt.legend()
    plt.show()
```

```python
for p in numbers:
    arr = np.load("C_"+str(m)+'order_'+(p)+".npy")
    c = []
    for i in range(0, len(arr)-1):
        t = arr[i+1] - arr[i]
        t = np.sum(t)
        t = t**2
        c.append(t)
    #print(np.arange(len(c)))
    #plt.plot(np.arange(len(c)), c , 'ro')
    plt.stem(np.arange(len(c)),c , "ro" , label = "l = "+p )
    plt.yscale("log")
    plt.legend()
    plt.show()
```

```python
for p in numbers:
    l_norm = []
    arr = np.load("C_"+str(m)+'order_'+(p)+".npy" )
    for i in arr:
        l_norm.append((i[0] - i[1] - i[2]-i[3])**2)
    plt.stem(np.arange(len(l_norm)) , l_norm ,  label = 'l = '+p)
    plt.yscale('log')
    plt.legend()
    plt.show()
```
